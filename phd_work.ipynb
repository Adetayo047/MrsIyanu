```python
# Function to convert year and week to date
def year_week_to_date(year, week):
    start_of_week = datetime.strptime(f'{year}-W{week}-1', '%Y-W%W-%w')
    return start_of_week

# Convert "Year" and "Week" columns to a new "Date" column
AllState['Date'] = AllState.apply(lambda row: year_week_to_date(row['Year'], row['Week']), axis=1)

# Display the DataFrame with the new "Date" column
AllState
```
