##Reading tthe data in Bauchi state

```python
alkeleri = pd.read_excel('Bauchi/alkeleri.xlsx')
Bauchi = pd.read_excel('Bauchi/Bauchi.xlsx')
Bogoro = pd.read_excel('Bauchi/Bogoro.xlsx')
Dass = pd.read_excel('Bauchi/Dass.xlsx')
TafawaBalewa = pd.read_excel('Bauchi/Tafawa-Balewa.xlsx')
Toro = pd.read_excel('Bauchi/Toro.xlsx')
Zaki = pd.read_excel('Bauchi/Zaki.xlsx')



display(alkeleri.head(),Bauchi.head(),Bogoro.head(),Dass.head(),TafawaBalewa.head(),Toro.head(),Zaki.head())
```

```python
#Renaming Columns
alkeleri.rename(columns={'Preciptation':'Precipitation'}, inplace = True)
Bauchi.rename(columns={'precipitation':'Precipitation',	'relative humidity':'Relative Humidity','Temp':'Temperature'}, inplace = True)
Bogoro.rename(columns={'precipitation':'Precipitation',	'Relative Humidity':'Relative Humidity','temperature':'Temperature'}, inplace =True)
Toro.rename(columns={'RelativeHumidity':'Relative Humidity'}, inplace = True)

display(alkeleri.head(),Bauchi.head(),Bogoro.head(),Dass.head(),TafawaBalewa.head(),Toro.head(),Zaki.head())
```

```python
# Assuming you have a list of dataframes called dataframes_list
dataframes_list = [alkeleri, Bauchi, Bogoro, Dass,TafawaBalewa,Toro,Zaki]

# Define the desired column order
desired_columns_order = ['LGA', 'Year', 'Week', 'variable','Temperature','Preciptation','Relative Humidity']

# Rearrange columns for all dataframes
for df in dataframes_list:
    df = df.reindex(columns=desired_columns_order)

dataframes_list
# display(alkeleri.head(),Bauchi.head(),Bogoro.head(),Dass.head(),TafawaBalewa.head(),Toro.head(),Zaki.head())

# Concatenate all of the DataFrames together into a single DataFrame
Bauchicomb = pd.concat(dataframes_list, ignore_index=True)

Bauchicomb= Bauchicomb[['LGA','Year','Week','variable','Temperature','Relative Humidity','Precipitation']]

Bauchicomb.head()
```
